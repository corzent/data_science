```python
import nltk
from urllib import request  
nltk.download('punkt')
```

```python
url = 'https://gutenberg.org/cache/epub/514/pg514.txt'
```

```python
response = request.urlopen(url)
raw = response.read().decode('utf-8')
type(raw)
```

```python
tokens = nltk.tokenize.word_tokenize(raw)
type(tokens)
```

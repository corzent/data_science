# **Метод наивного Байеса**
Теорема Байеса используется для вычисления вероятности того, что сообщение является спамом, основываясь на наличии определенных слов или других признаков.

**Наивное предположение**: Наивное Байесовское предположение заключается в том, что все признаки (слова в тексте) независимы друг от друга. Это позволяет упростить вычисления, хотя в реальной жизни слова обычно взаимосвязаны.

**Обучение модели**: Для того чтобы модель могла определять, является ли новое сообщение спамом, она должна сначала "обучиться" на размеченном наборе данных (где сообщения уже классифицированы как спам или не спам). Во время обучения модель рассчитывает:

- Вероятность появления каждого слова в спам-сообщениях.
- Вероятность появления каждого слова в не спам-сообщениях.

**Классификация новых сообщений**: Когда появляется новое сообщение, модель подсчитывает вероятность того, что оно является спамом, на основе встречаемых в нем слов. Сообщение будет классифицировано как "спам" или "не спам" в зависимости от того, какая вероятность окажется выше.

P(S|сообщение) = P(S) * P(слово1|S) - P(слово2|S) - .. P(слово n|S)

- веротность того что любое сообщение в наборе данных являестя спамом или P(S)
- вероятности того что каждое слово в сообщении встречается в спаме или P(слово|S)

P(S) - это доля сообщений в наборе данных, которые являются словами

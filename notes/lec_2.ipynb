В Котором модель определяет взаимосвязь между зависимым и независимыми значениями оно позволяет нам понять как какое-то значение которое зависит от какого-то переменного изменяется соответсвенно когда мы изменяем это значение 

Например используется чтобы определить возраст температуру 


Существуют типы регрессий
* Линейная регрессия 
* Логистическая регрессия
* Полиномиальная регрессия
* Поддерживая векторная регрессия 

**Линейная регрессия** - показывает линейную зависимость между независимым значением (ось Х) и зависимым значением (ось У)
Если используем только один входной параметр то называется - простым, а  если более одного - то множественная линейная регрессия
